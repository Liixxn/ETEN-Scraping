```python
import requests
from bs4 import BeautifulSoup

header = {
    'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10.15; rv:87.0) Gecko/20100101 Firefox/87.0',
    'Accept-Language': 'es'
}

```

```python
url_carrefour = "https://www.carrefour.es/supermercado/ofertas/N-177ap79Zwhajzd?No=0&Nr%3DAND%28product.shopCodes%3A004320%2Cproduct.salepointWithActivePrice_004320%3A1%2COR%28product.siteId%3AbasicSite%29%29OR%29=&prtId=cat20002"
page_carrefour_ofertas = requests.get(url_carrefour, headers=header)
soup_carrefour_ofertas = BeautifulSoup(page_carrefour_ofertas.content, 'html.parser')

print(soup_carrefour_ofertas)

```

```python
product_cards = soup_carrefour_ofertas.find_all('article', {'class': 'product-card-item'})

print(product_cards)

for card in product_cards:
    
    badge = card.find('div', {'class': 'bg-promocion-copy'})
    
    if (badge is None) :
        continue

    name = card.find('p', {'class': 'title-product'}).text.strip()
    
    price = card.find('span', {'class': 'price-less'}).text.strip()
    
    image_url = card.find('img', {'class': 'js-gap-product-click-super'})['src']
    
    print('Product Name:', name)
    print('Price:', price)
    print('Image URL:', image_url)

"""
flecha = soup_pag.find('span', {'class': 'pagination__previous pagination__previous--disabled icon-left-arrow-thin'})


if flecha is None
    continue

last_page_link = soup.find('a', {'class': 'pagination__item'})
total_pages = int(last_page_link['href'].split('/')[-1])
"""

```
