```python
# Librerias
import requests
from bs4 import BeautifulSoup
```

```python
# Headers para poder acceder a la pagina
header = {
    'User-Agent': 'Chrome 108.0.5359.125',
    'Accept-Language': 'es'
}
```

```python
# Se realiza una peticion a la pagina web de recetas online
url_corteingles = "https://www.alcampo.es/compra-online/ofertas-alimentacion/c/W04012023"
pag_corteingles = requests.get(url_corteingles, headers = header)
soup_corteingles = BeautifulSoup(pag_corteingles.content, 'html.parser')
```
